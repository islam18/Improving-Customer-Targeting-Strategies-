```python
import pandas as pd
import matplotlib.pyplot as plt

# First Approach data setup
data = {
    'Model': [
        'Original Dataset',
        'Balanced Dataset',
        'Without Preprocessing',
        'Without Undersampling',
        'Without Clustering'
    ],
    'F1_score': [0.36, 0.82, 0.77, 0.42, 0.31],
    'ROC AUC': [0.62, 0.87, 0.80, 0.65, 0.54]
}

df = pd.DataFrame(data)
df.set_index('Model', inplace=True)  # Set the 'Model' column as the index for easy plotting

# Create a figure with a specific size
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # Create 1 row and 2 columns for subplots
fig.suptitle('Ablation Study', fontsize=16)

# Define the metrics to plot next to each other
metrics = ['F1_score', 'ROC AUC']
colors = ['blue', 'purple']  # Colors for the histograms

# Plotting each specified metric in its own subplot
for i, metric in enumerate(metrics):
    axs[i].barh(df.index, df[metric], color=colors[i])
    axs[i].set_xlabel(metric)
    axs[i].set_title(f'Histogram of {metric}')

# Adjust the layout to ensure the plots do not overlap and everything is clearly visible
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
